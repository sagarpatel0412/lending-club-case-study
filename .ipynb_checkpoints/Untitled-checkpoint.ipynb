```python
#null thresold
threshold = 25000
```

```python
# indentify all column whose null count exceed more than null thresold
drop_column = lending_loan_data.columns[lending_loan_data.isnull().sum() > threshold]
drop_column
```

```python
#drop all columns more null thresold
lending_loan_data=lending_loan_data.drop(columns=drop_column)
```

```python
lending_loan_data.isnull().sum()
```

```python
#lets see if any other columns need to be cleared by analysing upper output
#url
#desc
#delinq_2yrs 
#emp_title
#last_pymnt_d
#last_credit_pull_d
#title
#emp_title
#total_pymnt
#total_pymnt_inv
#total_rec_prncp
#total_rec_int
#total_rec_late_fee
#recoveries
#collection_recovery_fee
#last_pymnt_amnt
```

```python
lending_loan_data.drop(['url','desc','delinq_2yrs','emp_title','last_pymnt_d','last_credit_pull_d','title','emp_title','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_amnt',],axis=1,inplace=True)
```

```python
lending_loan_data.isnull().sum()
```

```python
#Cleaning data within columns 1) cleaning term
lending_loan_data.term.value_counts()
```

```python
lending_loan_data.term = lending_loan_data.term.apply(lambda x : int(x.replace(" months", "")))
```

```python
lending_loan_data.term.value_counts()
```

```python
lending_loan_data.nunique()
```

```python
lending_loan_data.delinq_amnt.value_counts()
```

```python
#now we can remove column by checking what column with 1 unique values serve purpose
lending_loan_data=lending_loan_data.loc[:,lending_loan_data.nunique()>1]
```

```python
lending_loan_data.isnull().sum()
```

```python
print("List of column in dataframe \n",lending_loan_data.columns)
```

```python
#remove percentage sign from int_rate column
lending_loan_data.int_rate
```

```python
lending_loan_data.int_rate = lending_loan_data.int_rate.apply(lambda x : float(x.replace("%", "")))
```

```python
lending_loan_data.int_rate
```

```python
#remove percentage sign from int_rate column
lending_loan_data.revol_util
```

```python
lending_loan_data.revol_util= lending_loan_data.revol_util.apply(lambda x : float(str(x).replace("%", "")))
```

```python
lending_loan_data.revol_util
```

```python
#removing unnecesary string or text from emp_length 
lending_loan_data.emp_length=lending_loan_data.emp_length.apply(lambda x : str(x).replace(" years", ""))
lending_loan_data.emp_length=lending_loan_data.emp_length.apply(lambda x : str(x).replace(" year", ""))
lending_loan_data.emp_length=lending_loan_data.emp_length.apply(lambda x : str(x).replace("10+", "10"))
lending_loan_data.emp_length=lending_loan_data.emp_length.apply(lambda x : str(x).replace("< 1", "0"))
```

```python
lending_loan_data.emp_length.value_counts()
```

```python
#lets remove nan values from emp_length column
lending_loan_data['emp_length'] = lending_loan_data['emp_length'].replace(['nan', ''], float('nan'))
lending_loan_data = lending_loan_data.dropna(subset=['emp_length'])
```

```python
lending_loan_data.emp_length.value_counts()
```

```python
#lets insert 0 values in pub_rec_bankruptcies in place of NAN
lending_loan_data.pub_rec_bankruptcies.fillna(0,inplace=True)
```

```python
#checking shape of dataframe
lending_loan_data.shape
```

```python
lending_loan_data.isnull().sum()
```

```python
#replace null values with 0 in revol_util column
lending_loan_data.revol_util.value_counts()
```

```python
lending_loan_data.revol_util.fillna(0,inplace=True)
```

```python
lending_loan_data.isnull().sum()
```

```python
 #date is in mm-yy format splitting into different column
lending_loan_data.issue_d = pd.to_datetime(lending_loan_data.issue_d, format='%b-%y')
lending_loan_data['issue_d_month']=pd.DatetimeIndex(lending_loan_data.issue_d).month
lending_loan_data['issue_d_year']=pd.DatetimeIndex(lending_loan_data.issue_d).year
```

```python
lending_loan_data.earliest_cr_line = pd.to_datetime(lending_loan_data.earliest_cr_line, format='%b-%y')
lending_loan_data['earliest_cr_line_month']=pd.DatetimeIndex(lending_loan_data.earliest_cr_line).month
lending_loan_data['earliest_cr_line_year']=pd.DatetimeIndex(lending_loan_data.earliest_cr_line).year
```
