# Univariate Analysis

```python
#lets start with univariate analysis

## want to analysis loan status how much they have payed

plt.figure(figsize=(12, 12))
ax = sns.countplot(data=lending_loan_data, x='loan_status', order=lending_loan_data.loan_status.value_counts().index)

total = len(lending_loan_data)
for p in ax.patches:
    count = p.get_height()
    percentage = (count / total) * 100
    ax.annotate(
        f'{count:.0f}\n({percentage:.1f}%)',
        (p.get_x() + p.get_width() / 2., count),
        ha='center', va='center', xytext=(0, 5), textcoords='offset points'
    )

plt.title('Loan Status Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
#checking shape
lending_loan_data.shape
```

```python
# analysis are number of defaulter are less then fully paid amount
```

# Bivariate analysis

```python
#plot for grade vs loan status

plt.figure(figsize=(24, 14))
ax = sns.countplot(data=lending_loan_data, x='grade', hue='loan_status')


for p in ax.patches:
    count = p.get_height()
    if count > 0:
        ax.annotate(
            f'{count:.0f}',
            (p.get_x() + p.get_width() / 2., count),
            ha='center', va='center', xytext=(0, 5), textcoords='offset points'
        )


plt.xlabel('Grade')
plt.ylabel('Count')
plt.title('Distribution of Grade For Loan Status', fontsize=12)
plt.show()
```

```python
# Distribution is high for grad 'b' in loan status for both 'fully paid' and 'charged off'
```

```python
# plot for purpose and loan status

plt.figure(figsize=(12, 6))
ax = sns.countplot(data=lending_loan_data, x='purpose', hue='loan_status', palette="rocket")


for p in ax.patches:
    count = p.get_height()
    if count > 0:
        ax.annotate(
            f'{int(count)}',
            (p.get_x() + p.get_width() / 2., count),
            ha='center', va='center', xytext=(0, 5), textcoords='offset points'
        )

plt.xlabel('Purpose')
plt.ylabel('Count')
plt.title('Distribution of Loan Taken for Different Purposes', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#debt consolidation is higher peak of purpose with fully paid and defaulters
```

```python
# this graph shows tern and loan status

plt.figure(figsize=(24, 14))
ax = sns.countplot(data=lending_loan_data, x='sub_grade', hue='loan_status')


for p in ax.patches:
    count = p.get_height()
    if count > 0:
        ax.annotate(
            f'{count:.0f}',
            (p.get_x() + p.get_width() / 2., count),
            ha='center', va='center', xytext=(0, 5), textcoords='offset points'
        )


plt.xlabel('Sub Grade')
plt.ylabel('Count')
plt.title('Distribution of Sub Grade For Loan Status', fontsize=12)
plt.show()
```

```python
#distribution for loan per issue year
plt.figure(figsize=(11, 5))
ax = sns.countplot(data=lending_loan_data, x='issue_d_year', hue='loan_status')

for p in ax.patches:
    count = p.get_height()
    if count > 0:
        ax.annotate(
            f'{p.get_height():.0f}',
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', xytext=(0, 5), textcoords='offset points'
        )


plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Year For Loan Status', fontsize=12)
plt.show()
```
